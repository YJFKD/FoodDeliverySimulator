```python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn import linear_model
import matplotlib.pyplot as plt
```

# Instance_small_1

## Read data

```python
instance_small_1_data = pd.read_csv("../Data/Training_testing_data/Instance_small_1/output_data.csv", index_col=False)
instance_small_1_data = instance_small_1_data[['time_period_id', 'number_of_driver_index', 'num_orders', 'overdue', 'nearby_restaurant', 'restaurant_id', 'delivery_time']]
instance_small_1_data['time_period_id'] = instance_small_1_data['time_period_id'] - 3
instance_small_1_data['time_period_id'] = instance_small_1_data['time_period_id'].astype(str)
instance_small_1_data['restaurant_id'] = instance_small_1_data['restaurant_id'].astype(str)
instance_small_1_data
```

```python
instance_small_1_data.plot.scatter(x='number_of_driver_index', y='delivery_time')
```

```python
sns.pairplot(instance_small_1_data,x_vars=['number_of_driver_index'],y_vars=['delivery_time'],hue='restaurant_id', height=8, kind='reg')
```

```python
sns.lmplot(x="number_of_driver_index", y="delivery_time", data=instance_small_1_data[instance_small_1_data['restaurant_id'] == '2'])
```

## Clean data

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

Transformer = ColumnTransformer(transformers=[
    ('onehot', OneHotEncoder(), ['time_period_id', 'restaurant_id']),
    ('scale', StandardScaler(), ['num_orders', 'overdue', 'nearby_restaurant'])
], remainder='passthrough')

X = Transformer.fit_transform(instance_small_1_data.iloc[:,:-1])
y = instance_small_1_data.iloc[:,-1].values
```

```python
X, y
```

```python
X[0], X[1]
```

## Fit the model

### Linear Regression

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
regressor = LinearRegression()
regressor.fit(X_train, y_train)

print(regressor.coef_, regressor.intercept_)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

### Ridge regression

```python
regressor_ridge = linear_model.Ridge(alpha=.5)
regressor_ridge.fit(X_train, y_train)

print(regressor_ridge.coef_, regressor_ridge.intercept_)
```

```python
y_pred_ridge = regressor_ridge.predict(X_test)
y_pred_ridge
```

```python
df_ridge = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df_ridge
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred_ridge))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred_ridge))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred_ridge)))
print('R^2 score:', metrics.r2_score(y_test, y_pred_ridge))
print('Mean Absolute Percentage Error:', metrics.mean_absolute_percentage_error(y_test, y_pred_ridge))
```

## Evaluation

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R^2 score:', metrics.r2_score(y_test, y_pred))
print('Mean Absolute Percentage Error:', metrics.mean_absolute_percentage_error(y_test, y_pred))
```

# Instance_small_2

```python
instance_small_2_data = pd.read_csv("../Data/Training_testing_data/Instance_small_2/output_data.csv", index_col=False)
instance_small_2_data = instance_small_2_data[['time_period_id', 'number_of_driver_index', 'num_orders', 'overdue', 'nearby_restaurant', 'restaurant_id', 'delivery_time']]
instance_small_2_data['time_period_id'] = instance_small_2_data['time_period_id'] - 3
instance_small_2_data['time_period_id'] = instance_small_2_data['time_period_id'].astype(str)
instance_small_2_data['restaurant_id'] = instance_small_2_data['restaurant_id'].astype(str)
# instance_small_2_data.drop(['overdue'], axis=1, inplace=True)
instance_small_2_data
```

```python
instance_small_2_data.plot.scatter(x='number_of_driver_index', y='delivery_time')
```

```python
sns.pairplot(instance_small_2_data,x_vars=['number_of_driver_index'],y_vars=['delivery_time'],hue='restaurant_id', height=8)
```

```python
sns.lmplot(x="number_of_driver_index", y="delivery_time", data=instance_small_2_data[instance_small_2_data['restaurant_id'] == '2'])
```

```python
Transformer = ColumnTransformer(transformers=[
    ('onehot', OneHotEncoder(), ['time_period_id', 'restaurant_id']),
    ('scale', StandardScaler(), ['num_orders', 'overdue', 'nearby_restaurant'])
], remainder='passthrough')

X_2 = Transformer.fit_transform(instance_small_2_data.iloc[:,:-1])
y_2 = instance_small_2_data.iloc[:,-1].values

X_2, y_2
```

```python
X_2[0]
```

```python
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X_2, y_2, test_size=0.2, random_state=0)
regressor_2 = LinearRegression()
regressor_2.fit(X_train_2, y_train_2)

print(regressor_2.coef_, regressor_2.intercept_)
```

```python
y_pred_2 = regressor_2.predict(X_test_2)
y_pred_2
```

```python
df_2 = pd.DataFrame({'Actual': y_test_2, 'Predicted': y_pred_2})
df_2
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test_2, y_pred_2))
print('Mean Squared Error:', metrics.mean_squared_error(y_test_2, y_pred_2))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test_2, y_pred_2)))
print('R^2 score:', metrics.r2_score(y_test_2, y_pred_2))
print('Mean Absolute Percentage Error:', metrics.mean_absolute_percentage_error(y_test_2, y_pred_2))
```

```python
regressor_2_ridge = linear_model.Ridge(alpha=.5)
regressor_2_ridge.fit(X_train_2, y_train_2)

print(regressor_2_ridge.coef_, regressor_2_ridge.intercept_)
```

```python
y_pred_2_ridge = regressor_2_ridge.predict(X_test_2)
y_pred_2_ridge
```

```python
df_2_ridge = pd.DataFrame({'Actual': y_test_2, 'Predicted': y_pred_2_ridge})
df_2_ridge
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test_2, y_pred_2_ridge))
print('Mean Squared Error:', metrics.mean_squared_error(y_test_2, y_pred_2_ridge))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test_2, y_pred_2_ridge)))
print('R^2 score:', metrics.r2_score(y_test_2, y_pred_2_ridge))
print('Mean Absolute Percentage Error:', metrics.mean_absolute_percentage_error(y_test_2, y_pred_2_ridge))
```

## Advanced model

### Decision tree

```python
from sklearn.tree import DecisionTreeRegressor

regressor_2_tree = DecisionTreeRegressor(max_depth=8)
regressor_2_tree.fit(X_train_2, y_train_2)
y_pred_2_tree = regressor_2_tree.predict(X_test_2)
print(y_pred_2_tree)
```
