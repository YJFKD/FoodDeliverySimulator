```python
import pandas as pd
import numpy as np
import json
```

```python
with open('Food_delivery_simulator/Algorithm/order_update_history.json') as file:
    data = json.load(file)
    print(type(data))
```

```python
data
```

```python
column_list = ['order_id', 'creation_time', 'pickup_time', 'actual_completion_time', 'committed_completion_time']
result = pd.DataFrame(columns=column_list)

for key, value in data.items():
    order = key
    order_creation_time = value[1]["update_time"]
    order_pickup_time = value[2]["update_time"]
    order_actual_completion_time = value[3]["update_time"]
    order_comitted_completion_time = value[3]["committed_completion_time"]
    df = pd.DataFrame([[order, order_creation_time, order_pickup_time, order_actual_completion_time, order_comitted_completion_time]], columns=column_list)
    result = result.append(df, ignore_index=True)

result    
```

```python
result[['creation_time', 'pickup_time', 'actual_completion_time', 'committed_completion_time']] = \
result[['creation_time', 'pickup_time', 'actual_completion_time', 'committed_completion_time']].apply(pd.to_datetime)
result
```

```python
result['overdue'] = (result['committed_completion_time'].values.astype(np.int64) - result['actual_completion_time'].values.astype(np.int64)) // 10 ** 9
result
```

```python
overdue_df = result.loc[result['overdue'] < 0]
overdue_df
```

```python
overdue_df.sort_values(by=['overdue'])
```

```python
overdue_df.loc[overdue_df['overdue'] < -1000].sort_values(by=['overdue'])
```
